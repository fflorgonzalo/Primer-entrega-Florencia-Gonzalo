# CENTROS

```python
# df_apellidos = pd.read_csv("Desktop/FLOR/CODER - SQL/apellidos comunes.csv")
df_apellidos = pd.read_csv("apellidos comunes.csv")
df_apellidos
```

```python
lista_apellidos = df_apellidos['Apellidos'].tolist()
```

```python
#df_nombres = pd.read_csv("Desktop/FLOR/CODER - SQL/nombres comunes.csv")
df_nombres = pd.read_csv("nombres comunes.csv")
df_nombres
```

```python
df_nombres_male = df_nombres['Male Names']
df_nombres_female = df_nombres['Female Names']
df_nombres_male
df_nombres_female
```

```python
df_nombres_unidos = pd.concat([df_nombres_male, df_nombres_female], axis=0)
df_nombres_unidos
```

```python
df_nombres_mezclados = sklearn.utils.shuffle(df_nombres_unidos).reset_index(drop=True)
df_nombres_mezclados
```

```python
centros = pd.DataFrame()
centros['id_centro'] = 'id_centro'
centros['nombre'] = df_nombres_mezclados
centros['apellido'] = 'apellido'
```

```python
indice_centros = []
for i in range (1,2001,1):
    indice_centros.append(i)
```

```python
centros['id_centro'] = indice_centros
centros
```

```python
apellidos_aleatorios = centros['apellido'].apply(lambda x: random.choice(lista_apellidos))
apellidos_aleatorios
```

```python
centros['apellido'] = apellidos_aleatorios
centros
```

```python
centros['nombre'] = centros['nombre'].str.lower()
centros['apellido'] = centros['apellido'].str.lower()
centros
```

```python
df_soporte = pd.DataFrame()
df_soporte['mail_incompleto'] = centros[['nombre', 'apellido']].apply(''.join, axis=1)
df_soporte['@gmail'] = '@gmail.com'
df_soporte['mail'] =  df_soporte[['mail_incompleto', '@gmail']].apply(''.join, axis=1)
df_soporte
```

```python
centros['mail'] = df_soporte['mail']
centros
```

```python
centros['telefono'] = 'telefono'
centros['telefono'] = centros['telefono'].apply(lambda x: random.randrange(1000000000,10000000000,1))
```

```python
centros.to_csv("Desktop/FLOR/CODER - SQL/Creacion de tablas/centros.csv")
```
